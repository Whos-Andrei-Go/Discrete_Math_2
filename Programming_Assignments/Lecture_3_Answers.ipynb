```python
import matplotlib.pyplot as plt
import networkx as nx

network = nx.Graph()
network.add_nodes_from(['A','B','C'], bipartite=0) # Add the node attribute "bipartite"
network.add_nodes_from(['1','2', '3'], bipartite=1)
network.add_edges_from([('A', '1'), ('A', '2'), ('A', '3'),
                              ('B', '1'), ('B', '2'), ('B', '3'),
                              ('C', '1'), ('C', '2'), ('C', '3')])

top = nx.bipartite.sets(network)[0]
pos = nx.bipartite_layout(network, top)
nx.draw(network, pos=pos, with_labels=True, node_color=['pink','magenta','red','deepskyblue','cyan', 'green'])
plt.show()
```

The graph depicted is a bipartite graph with 3 nodes in one partition and 3 nodes in another. Nodes 'A', 'B', and 'C' belong to the first partition, while nodes '1', '2', and '3' belong to the second partition. Each node in one partition is connected to every node in the other partition, resulting in a complete bipartite graph. Specifically, nodes 'A', 'B', and 'C' are connected to nodes '1', '2', and '3', respectively. The nodes are positioned using a bipartite layout, where nodes in the same partition are aligned vertically. This graph serves as an example of a bipartite graph, demonstrating how nodes can be grouped into two disjoint sets with edges only between nodes of different sets.

$\textbf{The Handshake Theorem}$
---

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a complete graph with 2 nodes
G = nx.complete_graph(4)

# Seed for reproducible layout
seed = 57
pos = nx.spring_layout(G, seed=seed)

# Draw the graph
nx.draw(G, pos=pos, with_labels=True)
plt.show()
```

$deg(v_0)$ = 3

$deg(v_1)$ = 3

$deg(v_2)$ = 3

$deg(v_3)$ = 3

Total Edges: 6

$deg(v_0)$ + $deg(v_1)$ + $deg(v_2)$ + $deg(v_3)$ = 12

Edges * 2 = Total Degree

6 * 2 = 12

Since the total degree of the graph is even, the graph exists.

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a graph for each concept
walk = nx.Graph()
walk.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 5), (5, 1), (1, 3)])

trail = nx.Graph()
trail.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 5), (5, 3)])

path = nx.Graph()
path.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 5), (5, 3)])

closed_walk = nx.Graph()
closed_walk.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1)])

circuit = nx.Graph()
circuit.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1), (1, 3)])

simple_circuit = nx.Graph()
simple_circuit.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1)])

# Define the paths
walk_path = [(1, 2), (2, 3), (3, 4), (4, 5), (5, 1), (1, 3)]
trail_path = [(1, 2), (2, 3), (3, 4), (4, 5), (5, 3)]
path_path = [(1, 2), (2, 3), (3, 4), (4, 5)]
closed_walk_path = [(1, 2), (2, 3), (3, 4), (4, 1)]
circuit_path = [(1, 2), (2, 3), (3, 4), (4, 1)]
simple_circuit_path = [(1, 2), (2, 3), (3, 4), (4, 1)]

# Plot each graph
plt.figure(figsize=(15, 10))

# Walk
plt.subplot(231)
nx.draw(walk, pos=nx.circular_layout(walk), with_labels=True, node_size=500, node_color='lightblue')
nx.draw_networkx_edges(walk, pos=nx.circular_layout(walk), edgelist=walk_path, edge_color='green', width=2)
plt.title('Walk')

# Trail
plt.subplot(232)
nx.draw(trail, pos=nx.planar_layout(trail), with_labels=True, node_size=500, node_color='lightblue')
nx.draw_networkx_edges(trail, pos=nx.planar_layout(trail), edgelist=trail_path, edge_color='green', width=2)
plt.title('Trail')

# Path
plt.subplot(233)
nx.draw(path, pos=nx.spring_layout(path, seed=55), with_labels=True, node_size=500, node_color='lightblue')
nx.draw_networkx_edges(path, pos=nx.spring_layout(path, seed=55), edgelist=path_path, edge_color='green', width=2)
plt.title('Path')

# Closed Walk
plt.subplot(234)
nx.draw(closed_walk, pos=nx.spring_layout(closed_walk, seed = 56), with_labels=True, node_size=500, node_color='lightblue')
nx.draw_networkx_edges(closed_walk, pos=nx.spring_layout(closed_walk, seed = 56), edgelist=closed_walk_path, edge_color='green', width=2)
plt.title('Closed Walk')

# Circuit
plt.subplot(235)
nx.draw(circuit, pos=nx.spring_layout(circuit, seed = 57), with_labels=True, node_size=500, node_color='lightblue')
nx.draw_networkx_edges(circuit, pos=nx.spring_layout(circuit, seed = 57), edgelist=circuit_path, edge_color='green', width=2)
plt.title('Circuit')

# Simple Circuit
plt.subplot(236)
nx.draw(simple_circuit, pos=nx.spring_layout(simple_circuit, seed = 58), with_labels=True, node_size=500, node_color='lightblue')
nx.draw_networkx_edges(simple_circuit, pos=nx.spring_layout(simple_circuit, seed = 58), edgelist=simple_circuit_path, edge_color='green', width=2)
plt.title('Simple Circuit')

plt.tight_layout()
plt.show()
```

1. Walk:
A walk in a graph is a sequence of vertices and edges, where consecutive vertices are connected by edges. In the walk graph, a walk is depicted starting from vertex 1 and traversing through vertices 2, 3, 4, and 5, forming a closed loop back to vertex 1. The green edges highlight the path of the walk.

2. Trail:
A trail in a graph is a walk where no edge is repeated. In the trail graph, a trail is shown starting from vertex 1 and traversing through vertices 2, 3, 4, and 5. The green edges highlight the path of the trail.

3. Path:
A path in a graph is a trail where no vertex is repeated. In the path graph, a path is illustrated starting from vertex 1 and traversing through vertices 2, 3, 4, and 5. The green edges highlight the path of the trail.

4. Closed Walk:
A closed walk in a graph is a walk where the starting and ending vertices are the same. In the closed walk graph, a closed walk is demonstrated starting from vertex 1 and traversing through vertices 2, 3, and 4, forming a loop back to vertex 1. The green edges highlight the path of the closed walk.

5. Circuit:
A circuit in a graph is a closed walk where no edge is repeated, except for the starting and ending vertices. In the circuit graph, a circuit is shown starting and ending at vertex 1 and traversing through vertices 2, 3, and 4. The green edges highlight the path of the circuit.

6. Simple Circuit:
A simple circuit in a graph is a circuit where no vertex is repeated, except for the starting and ending vertices. In the simple circuit graph, a simple circuit is illustrated starting and ending at vertex 1 and traversing through vertices 2, 3, and 4. The green edges highlight the path of the simple circuit.

These graphs demonstrate different concepts in graph theory, including walks, trails, paths, closed walks, circuits, and simple circuits, each with their specific properties and characteristics.

$\textbf{The Travelling Salesman Problem}$
---

```python
import matplotlib.pyplot as plt
import networkx as nx

# Define the cities and their coordinates
cities = {'A': (0, 0), 'B': (1, 3), 'C': (4, 1), 'D': (5, 4), 'E': (2, 5)}

# Create a complete graph with distances as weights
G = nx.Graph()
for city1 in cities:
    for city2 in cities:
        if city1 != city2:
            distance = round(((cities[city1][0] - cities[city2][0]) ** 2 + (cities[city1][1] - cities[city2][1]) ** 2) ** 0.5)
            G.add_edge(city1, city2, weight=distance * 10)

# Find an optimal route for the TSP using the built-in function in networkx
optimal_route = nx.approximation.traveling_salesman_problem(G)

# Plot the graph with the optimal route
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G, seed = 0)  # Position nodes with a spring layout
nx.draw(G, pos, with_labels=True, node_size=500, node_color='lightblue')
nx.draw_networkx_edges(G, pos, edgelist=[(optimal_route[i], optimal_route[i+1]) for i in range(len(optimal_route)-1)] + [(optimal_route[-1], optimal_route[0])], edge_color='green', width=2)  # Add the last edge to complete the cycle

# Add distances to the edges
edge_labels = {(edge[0], edge[1]): str(weight) for edge, weight in nx.get_edge_attributes(G, 'weight').items()}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red')

plt.title("Traveling Salesman Problem")
plt.show()
```

Route:
$A \to B \to E \to D \to C \to A = 30 + 20 + 30 + 30 + 40 = 150$

The graph illustrates a solution to the Traveling Salesman Problem (TSP), where a salesman must visit a set of cities exactly once and return to the starting city, while minimizing the total distance traveled. Five cities, labeled A, B, C, D, and E, are positioned with their respective coordinates. The edges between cities represent the distances between them, with weights corresponding to the Euclidean distance.

A complete graph is constructed, where each city is connected to every other city. The green edges depict the optimal route, starting at city A and traversing through the other cities before returning to A, forming a closed loop.

Additionally, the distance of each edge is labeled in red to indicate the distance traveled along the route. This graph demonstrates a solution to the TSP, showing how the salesman can visit all cities exactly once while minimizing the total distance traveled.
